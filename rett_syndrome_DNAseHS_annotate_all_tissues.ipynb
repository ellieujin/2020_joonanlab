```python
# DHS peak filtering function  
# f :str   / example: '/home/titan/resources/DNase/E059-DNase.hotspot.fdr0.01.broad.bed'
# include this function at the loop
def DHS_filter(f, SNV_Indel):
    bed = hl.import_bed(f, contig_recoding={f'chr{i}':str(i) for i in [*range(1,23), 'X', 'Y']})
    
    bed = bed.annotate(new_interval = hl.liftover(bed.interval, 'GRCh38'))  
    bed = bed.filter(hl.is_defined(bed.new_interval))  
    bed = bed.key_by(interval = bed.new_interval)
    
    SNV_Indel = SNV_Indel.annotate(tissue_name = hl.cond(hl.is_defined(bed[SNV_Indel.locus]), 1, 0))
    
    return SNV_Indel
```
